## Sigmoid family function

### Standard Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def standard_sigmoid():
    x = np.linspace(-10, 10, 100)
    y = 1 / (1 + np.exp(-x))
    plt.plot(x, y)
    plt.show()

standard_sigmoid()
```

### 3.2.1 Shifted and Scaled Sigmoid

Activation defined as:

$f(z) = \sigma(a(z-b)) = \frac{1}{1 + \text{exp}(-a(z-b))}$

```python
def shifted_scaled_sigmoid(a=0.02, b=600):
    x = np.linspace(-0, 1000, 100)
    y = 1 / (1 + np.exp(-a * (x - b)))
    plt.plot(x, y)
    plt.show()

shifted_scaled_sigmoid()
```

### 3.2.2 Variant sigmoid function (VSF)

Activation defined as:

$f(z) = a\sigma(bz) - c = \frac{a}{1+\text{exp}(-bz)}-c$

```python
def variant_sigmoid(a=0.02, b=10, c=10):
    x = np.linspace(-10, 10, 100)
    y = a / (1 + np.exp(-b*x)) -c
    plt.plot(x, y)

for a in [0.02, 0.1, 0.5]:
    for b in [10, 20, 50]:
        for c in [0, 5, 10]:
            variant_sigmoid(a, b, c)
plt.show()
```

## 3.2.3 Scaled hyperbolic tangent

Activation defined as:

$f(z) = a\tanh(bz)$



```python
def scaled_hyperbolic_tangent(a=0.5, b=2):
    x = np.linspace(-10, 10, 100)
    y = a * np.tanh(b * x)
    
    plt.plot(x, y, label=f"a={a}, b={b}")
    

for a in range(1, 10,3):
    for b in range(1, 10,3):
        scaled_hyperbolic_tangent(a,b)
plt.legend()
plt.show()        
```

### 3.2.4 Arctangent

Activation defined as:

$f(z) = \text{arctan}(z)$

```python
def arctan():
    x = np.linspace(-10, 10, 100)
    y = np.arctan(x)
    plt.plot(x, y)
    plt.show()

arctan()

```

### 3.2.5 Sigmoid-Algebraic activation function

Activation defined as:

$f(z) = \frac{1}{1 + \text{exp}(-  \frac{z(1+a|z|)}{1+|z|(a+a|z|)}   )}$

```python
def sigmoid_algebraic_activation(a):
    x = np.linspace(-10, 10, 100)
    exp_part = (x*(1+a*np.abs(x))) / (1+np.abs(x)*(1+a*np.abs(x)))
    y = 1 / (1 + np.exp(-exp_part))
    plt.plot(x, y)

for a in range(1,10):
    sigmoid_algebraic_activation(a)
plt.show()
```

### 3.2.6 Triple-state sigmoid

Activation defined as:

$f(z) = \frac{1}{1 + \text{exp}(-z)} ( \frac{1}{1+\text{exp}(-z)} + \frac{1}{1+\text{exp}(-z + a)}  + \frac{1}{1+\text{exp}(-z + b)})  $

```python
def triple_state_sigmoid(a=10, b=10.9):
    x = np.linspace(-4, 15, 100)
    part_a = 1 / (1 + np.exp(-x))
    part_b = 1 / (1 + np.exp((-x + a)))
    part_c = 1 / (1 + np.exp((-x + b)))
    y = part_a*(part_a + part_b + part_c)    
    plt.plot(x, y, label=f"a={a}, b={b}")

for a in range(1, 10,3):
    for b in range(1, 10,2):
        triple_state_sigmoid(a,b)
plt.legend()
plt.show()
```

### 3.2.7 Improved logistic sigmoid

$f(z) = a(z-b) + \sigma(b), z \geq b$

$f(z) = \sigma(b), -b \lt z \lt b$

$f(z) = a(z+b) + \sigma(b), z \leq -b$

```python
def improved_logistic_sigmoid(a=10, b=9.9):
    x = np.linspace(-10, 10, 100)
    sigm = lambda x: 1 / (1 + np.exp(-x))

    y = np.zeros(len(x))

    for i in range(len(x)):
        
        if x[i] >= b:
            y[i] = a*(x[i]-b) + sigm(x[i])
        
        elif x[i] > -b and x[i] < b:
            y[i] = sigm(x[i])

        else:
            y[i] = a*(x[i]+b) + sigm(x[i])

    plt.plot(x, y, label=f"a={a}, b={b}")

for a in range(1, 10, 3):
    for b in range(1, 10,3):
        improved_logistic_sigmoid(a,b)
plt.legend()
plt.show()
```

### 3.2.8 Combination of the sigmoid and linear (SigLin)

Activation defined as:

$f(z) = \frac{1}{1 + \text{exp}(-z)} + az$

```python
def siglin(a):
    x = np.linspace(-10, 10, 100)
    sig_part = 1 / (1 + np.exp(-x))
    lin_part = a*x
    y = sig_part + lin_part
    plt.plot(x, y, label=f"a={a}")

for a in [0,0.05,0.1,0.15]:
    siglin(a)

plt.legend()
plt.show()
```

### 3.2.9 Penalized hyperbolic tangent

Activation defined as:

$f(z) = \tanh(z), z \geq 0$

$f(z) = \frac{\tanh(z)}{a}, z \lt 0$

```python
def penalized_hyperbolic_tangent(a):
    x = np.linspace(-10, 10, 100)
    
    y = np.zeros(len(x))

    for i in range(len(x)):
        if x[i] >= 0:
            y[i] = np.tanh(x[i])
        else:
            y[i] = np.tanh(x[i])/a


    plt.plot(x, y, label=f"a={a}")

for a in range(10):
    penalized_hyperbolic_tangent(a)
plt.legend()
plt.show()
```
